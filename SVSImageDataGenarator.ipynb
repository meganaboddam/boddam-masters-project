```python
from PIL import Image
import os
import glob
import numpy as np
```

```python
def crop(im, height, width):
    imgwidth, imgheight = im.size
    rows = np.int(imgheight/height)
    cols = np.int(imgwidth/width)
    for i in range(rows):
        for j in range(cols):
            # print (i,j)
            box = (j*width, i*height, (j+1)*width, (i+1)*height)
            yield im.crop(box)
```

```python
if __name__ == '__main__':
    
    imgdir = 'data/train/normal'
    outputdir = 'data/train/normal/normal1_tiles'
    basename = 'normal1.svs'
    tile_size = 256

    filename = os.path.join(imgdir, basename)
    filenum = 1
    print(f'name is {filename}')

    im = Image.open(filename).convert('RGB')
    #im = im.convert('RGB')
    imgwidth, imgheight = im.size
    print(('Image size is: %d x %d ' % (imgwidth, imgheight)))

    height = np.int(imgheight/40)
    width = np.int(imgwidth/10)

    start_num = 0
    for k, piece in enumerate(crop(im, height, width), start_num):
      if k > 100:
        break
      img = Image.new("RGB", (width, height), tile_size)
      img.paste(piece)
      #print("Tile mean value: ", np.mean(img))
      if (np.mean(img) >= 240):
        print("Image Section is all white") 
      else:
        path = os.path.join(outputdir, "tile%d_1%05d.tiff" % (int(k+1),filenum))
        img.save(path)
        print("Image section saved: ", path)
        os.rename(path, os.path.join(outputdir, "tile%d.1%05d" % (int(k+1), filenum)))

      

        
```
